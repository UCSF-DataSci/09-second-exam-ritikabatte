```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv('/Users/ritikabatte02/09-second-exam-ritikabatte/ms_data.csv')   
```

## Walking speed analysis:

```python
# Scatter plot of age vs walking_speed with Regression Line
sns.regplot(x='age', y='walking_speed', data=data, line_kws={'color': 'red'}) 
plt.title('Scatter plot of age vs walking_speed with Regression Line')
plt.show()
```

```python
#  Box plot of walking_speed by education_level
sns.boxplot(x='education_level', y='walking_speed', data=data)
plt.title('Box plot of walking_speed by education_level')
plt.show()
```

```python
# Line plot showing interaction of education_level and age on walking_speed
sns.lineplot(x='age', y='walking_speed', hue='education_level', data=data)  
plt.title('Line plot showing interaction of education_level and age on walking_speed')
plt.show()
```

## Cost analysis: 

```python
# # define cost per visit for each insurance type
insurance_costs = {
    'Basic': 80,
    'Premium': 130,
    'Platinum': 210
}

data = pd.DataFrame({
    'insurance_type': ['Basic', 'Premium', 'Platinum'],
    'cost': [80, 130, 210]
})

# Bar plot of mean costs by insurance type with error bars
sns.barplot(x='insurance_type', y='cost', data=data, ci='sd')
plt.title('Bar plot of mean costs by insurance types')
plt.show()
```

```python
# Box plot showing cost distributions by insurance type
sns.boxplot(x='insurance_type', y='cost', data=data)
plt.title('Box plot of cost distributions by insurance type')
plt.show()
```

## Combined visualizations: 

```python
data = pd.DataFrame({
    'age': [25, 30, 35, 40, 45, 50],
    'walking_speed': [1.2, 1.4, 1.3, 1.5, 1.4, 1.6],
    'education': ['High School', 'College', 'College', 'Graduate', 'Graduate', 'High School'],
    'insurance_type': ['Basic', 'Premium', 'Platinum', 'Basic', 'Premium', 'Platinum'],
    'date': pd.date_range(start='2023-01-01', periods=6, freq='ME')
})
```

```python
# Pair Plot of Key Variables
sns.pairplot(data, vars=['age', 'walking_speed'], hue='insurance_type')
plt.suptitle('Pair Plot of Age and Walking Speed by Insurance Type', y=1.02)
plt.show()
```

```python
# Facet by insurance type for age vs. walking speed
g = sns.FacetGrid(data, col='insurance_type')
g.map(sns.scatterplot, 'age', 'walking_speed')
plt.suptitle('Facet by Insurance Type for Age vs. Walking Speed', y=1.02)
plt.show()
```

```python
# Facet by education for age vs. walking speed
g = sns.FacetGrid(data, col='education')
g.map(sns.scatterplot, 'age', 'walking_speed')
plt.suptitle('Facet by Education for Age vs. Walking Speed', y=1.02)
plt.show()
```

```python
# Time Trends 
sns.lineplot(x='date', y='walking_speed', hue='insurance_type', data=data)
plt.title('Time Trends of Walking Speed by Insurance Type')
plt.show()
```
